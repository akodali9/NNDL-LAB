```python
import numpy as np
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

print("Input Given: \n", X)

y = np.array([0, 0, 0, 1])

print("Expected Output: \n", y)
```

```python
w = np.array([1, 1])  # weight
```

```python
theta = 1.5
```

```python
y_pred = []
```

```python
def Heaviside (net):
    if net >= theta:
        return 1
    else:
        return 0
```

```python
for i in range(X.shape[0]):
    x = X[i]
    print("x: ",x)
    weighted_sum= np.dot(x,w)
    print("Weighted sum ",weighted_sum)
    pred = Heaviside(weighted_sum)
    print(x[0], " AND ",x[1]," -> ",pred, "\n")
    y_pred.append(pred)
```

```python
# Printing the computed output
print(y_pred)
```

```python
import matplotlib.pyplot as plt
slope = -w[0]/w[1]
intercept = (theta)/w[1]
x_plane = np.linspace(-2,2,10)
y_plane = slope * x_plane + intercept
plt.title("Decision Boundary")
plt.scatter(X[:,0], X[:,1], c=y)
plt.plot(x_plane, y_plane, '-')
plt.xlim([-0.2, 1.2])
plt.ylim([-0.2, 1.2])
plt.show()
```
