```python
import numpy as np
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

print("Input Given: \n", X)

y = np.array([0, 1, 1, 1])

print("Expected Output: \n", y)
```

```python
w = np.array([1, 1])
```

```python
theta = 1
```

```python
y_pred = []
```

```python
def Heaviside(net):
    if net >= theta:
        return 1
    else:
        return 0
```

```python
for i in range(X.shape[0]):
    x = X[i]
    wtsum = np.dot(x, w)
    pred = Heaviside(wtsum)
    print(x[0], "OR", x[1], "->", pred)
    y_pred.append(pred)
```

```python
print("Computed Output: \n", y_pred)
```

```python
# plotting the decision boundary

import matplotlib.pyplot as plt

slope = -w[0]/w[1]  # slope of the line

intercept = (theta)/w[1]  # intercept of the line

# Return evenly spaced numbers over a specified␣
x_plane = np.linspace(-2, 2, 10)

# interval
y_plane = slope * x_plane + intercept  # values calculated from x_plane with␣

# given slope and intercept
plt.scatter(X[:, 0], X[:, 1], c=-y)  # A scatter plot of y vs. x with varying␣

# marker size and/or color
# Plot y_plane versus x_plane as lines␣
plt.plot(x_plane-0.1, y_plane-0.1, '-')

# and/or markers
plt.xlim([-0.2, 1.2])  # Get or set the x limits of the current axes

plt.ylim([-0.2, 1.2])  # Get or set the y limits of the current axes

plt.show()  # Display all open figures
```
